```python
# 

#
import matplotlib.pyplot as plt
import random

#
plt.figure(figsize=(20,8),dpi = 80)

#,11121
#15~18
x = range(60)
y_jingxian = [random.uniform(15,18) for i in x]#uniform
y_beijing = [random.uniform(3,8) for i in x]#uniform

#
plt.plot(x,y_jingxian,label = '')
plt.plot(x,y_beijing,label = '')

plt.rcParams['font.family']=['SimHei']

#
plt.xlabel('',fontsize = 20)
plt.ylabel('',fontsize = 20)
plt.title('11121',fontsize = 30)

# # x
x_tick_label = ["11{}".format(i) for i in x]
plt.xticks(x[::5],x_tick_label[::5])# [::5]

# # y
y_range = range(40)
plt.yticks(y_range[::5])#[::5]

# linestyle alpha
plt.grid(True,linestyle = ':',alpha=0.8)

# loc
plt.legend(loc = 'best')

#
plt.savefig('./11121.png')

#
plt.show()#show
```
