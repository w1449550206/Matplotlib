```python
#
import matplotlib.pyplot as plt
import random

#
plt.figure(figsize=(20,8),dpi = 80)

#,11121
#15~18
x = range(60)
y_jingxian = [random.uniform(15,18) for i in x]#uniform

plt.plot(x,y_jingxian)

plt.rcParams['font.family']=['SimHei']

#
plt.xlabel('',fontsize = 20)
plt.ylabel('',fontsize = 20)
plt.title('11121',fontsize = 20)

# # 
x_tick_label = ["11{}".format(i) for i in x]
plt.xticks(x,x_tick_label)

#
plt.savefig('./11121.png')

#
plt.show()#show
```
